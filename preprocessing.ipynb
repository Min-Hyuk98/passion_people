```python
import csv
import pandas as pd

data = pd.read_csv("00_musinsa_data_man_trans.csv")
data.columns = ['dt','style','top_img','top_tag','bottom_img','bottom_tag']


def pretreat(df, top_bottom):
    styles = []
    tags = []
    styles = df['style']
    if top_bottom == "top":
        tags = df['top_tag']
    else:
        tags = df['bottom_tag']
 
    combined_1 = []
    for i in tags:
        i = i.replace("Street / Hip Hop", "Street/HipHop")
        i = i.replace(" / ", "/")
        i = i.replace("[","")
        i = i.replace("]","")
        i = i.replace(",","")
        i = i.replace("'","")
        i = i.replace('"',"")
        combined_1.append(i.split(" "))
#     print(combined_1)
    
    combined_2 = []
    for i in combined_1:
        ttt = []
        for j in i:
            ttt.append(j.lower())
        combined_2.append(ttt)
#     print(combined_2)

    final_list = []
    for i in combined_2:
        ttmp_list = []
        for j in i:
            if j == '/' or j == '' or j == 'x' or j == 'v' or j == 'a' or j =='el' or j == 'and' or j == 'or':
                continue
            else:
                ttmp_list.append(j)
        final_list.append(ttmp_list)
#     print(final_list)
    
    for idx, i in enumerate(final_list):
        final_list[idx] = list(set(i)) 
    
#     print(final_list)
    return final_list
    

final_list = pretreat(data, "top")
data['top_tag'] = final_list

final_list = pretreat(data, 'bottom')
data['bottom_tag'] = final_list

data.to_csv("musinsa_data_man.csv", mode='w')
```
